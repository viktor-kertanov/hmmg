```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import permutations
```

```python
num_vertex = 40
vertex_diff = 2
x_start = 0
y_start_init = 0
vertex_height = 80

num_horizontals = 1
avg_horizontal_length = num_vertex * vertex_diff + 13

horizontal_x_start = -6
horizontal_y_start = 0

thetas = np.linspace(0, 2*np.pi, 10)
circle_factor = .1

for vertex in range(num_vertex):
    cur_x = x_start+vertex*vertex_diff + np.random.rand()*.01
    y_start = y_start_init-np.random.rand()*1.5
    y_stop = y_start_init + vertex_height+np.random.rand()*1.5
    for _ in range(np.random.randint(1,3)):
        plt.plot([cur_x, cur_x], [y_start, y_stop], 'k-', linewidth = .75)
        plt.plot(np.cos(thetas)*circle_factor+cur_x, np.sin(thetas)*circle_factor+y_start_init + vertex_height+2, 'k-', linewidth=.5)
        
        plt.plot([cur_x, cur_x], [y_stop, y_start], 'k-', linewidth = .75)
        plt.plot(np.cos(thetas)*circle_factor+cur_x, np.sin(thetas)*circle_factor-2, 'k-', linewidth=.5)

    if vertex == num_vertex-1:
        plt.plot([cur_x, cur_x], [y_start, y_stop], 'k-', linewidth = .75)
    
plt.axis('off')
plt.axis([-30, 120, -10, 90])
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(8.25*2, 6*2)
plt.savefig('am_verices.svg')
plt.show()

cur_y = horizontal_y_start
horizontal_diff = np.random.rand()*.2 + .2 + np.random.rand()*0.09
x_start_horizont = horizontal_x_start
x_stop_horizont = x_start_horizont + avg_horizontal_length

x_random_factor = 1.7

while cur_y < (y_start + vertex_height-1):
    cur_y = cur_y + horizontal_diff
    x_start_horizont = x_start_horizont + np.random.rand()*x_random_factor - x_random_factor/2
    x_stop_horizont = x_stop_horizont + np.random.rand()*x_random_factor - x_random_factor/2
    for _ in range(np.random.randint(1,3)):
        plt.plot([x_start_horizont, x_stop_horizont], [cur_y, cur_y], 'r-', linewidth=.5)
        plt.plot([x_stop_horizont, x_start_horizont], [cur_y, cur_y], 'r-', linewidth=.5)
    
    horizontal_diff = np.random.rand()*.4 + .35 + np.random.rand()*0.01

plt.axis('off')
plt.axis([-30, 120, -10, 90])
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(8.25*2, 6*2)
plt.savefig('am_horizontals.svg')
plt.show()
```
