```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import permutations, combinations
```

```python
thetas = np.linspace(0, 2*np.pi, 100)

# Dealing with circles
x_start = 0
y_start = 0

radius = 1
r_small = .1 #for points
x_gap = radius*2.5
y_gap = radius*2.5

row_num_circles = 5
col_num_circles = 5



for c in range(col_num_circles):
    for r in range(row_num_circles):
        plt.plot(np.cos(thetas)*radius + x_gap*r, np.sin(thetas)*radius + c*y_gap, 'k-', linewidth=1)

        # Dealing with random points 
        num_points = 8
        random_angles = np.random.choice(thetas, num_points, replace=False)
        random_points = [(np.cos(angle) * radius + x_gap * r, np.sin(angle) * radius + c * y_gap) for angle in random_angles]
        rand_combs = combinations(random_points, r=2)

        for point in random_points:
            plt.plot(point[0], point[1], 'ro')

        for combination in rand_combs:
            plt.plot([combination[0][0], combination[1][0]], [combination[0][1], combination[1][1]], 'r-')

plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(14, 14)
plt.show()
```
