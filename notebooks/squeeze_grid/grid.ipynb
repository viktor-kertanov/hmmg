```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# squeezing number, geometric

n = 40
counter = np.arange(n)
q = 1.05

first = 1
seq = first*q**(counter)
y = np.zeros(n)

y_min = 0
y_delta = .1
num_lines = 20

y_max = y_min + y_delta*(num_lines-1)

for i in np.arange(num_lines):
    plt.plot(seq, y, 'k-')
    y += y_delta

for x_idx in range(n):
    plt.plot([seq[x_idx], seq[x_idx]], [y_min, y_max], 'k-')


for i in np.arange(num_lines):
    plt.plot(-seq, y, 'k-')
    y += y_delta

for x_idx in range(n):
    plt.plot([-seq[x_idx], -seq[x_idx]], [y_min, y_max], 'k-')


plt.gcf().set_size_inches(10,8)
plt.show()
```

```python
# squeezing number, geometric

n = 30
counter = np.arange(n)
q = 1.05

first = 1
seq_half = first*q**(counter)-0.98
seq = np.append(seq_half, -seq_half)
y = np.zeros(n*2)

y_min = 0
y_delta = .3
num_lines = 20

y_max = y_min + y_delta*(num_lines-1)

for i in np.arange(num_lines):
    plt.plot(seq, y, 'r-')
    y += y_delta

for x_idx, x in enumerate(seq):
    plt.plot([x, x], [y_min, y_max], 'r-')

for y_idx, y in enumerate(np.arange(y_min, y_max, y_delta)):
    for x_idx, x in enumerate(seq[:-1]):
        if (x_idx % 2 == 0 and y_idx % 2 == 0) or (x_idx % 2 == 1 and y_idx % 2 == 1):
            next_x = seq[x_idx+1]
            hatch_num = 5
            hatch_y_delta = y_delta/hatch_num
            for hatch_idx in range(1, hatch_num):
                plt.plot([x, next_x], [y+hatch_y_delta*hatch_idx, y+hatch_y_delta*hatch_idx], 'k-')




plt.gcf().set_size_inches(15,10)
plt.gca().set_aspect('equal')
plt.show()
```

```python
# squeezing number, geometric

n = 35
counter = np.arange(n)
q = 1.05

first = 1
seq_half = first*q**(counter)-0.98
seq = np.sort(np.append(seq_half, -seq_half))
y = np.zeros(n*2)

y_min = 0
y_delta = .3
num_lines = 22

y_max = y_min + y_delta*(num_lines-1)

#horizontal main lines
for i in np.arange(num_lines):
    plt.plot(seq, y, 'b-')
    y += y_delta

for x_idx, x in enumerate(seq):
    plt.plot([x, x], [y_min, y_max], 'b-')

for y_idx, y in enumerate(np.arange(y_min, y_max, y_delta)):
    for x_idx, x in enumerate(seq[:-1]):
        if (x_idx % 2 == 0 and y_idx % 2 == 0) or (x_idx % 2 == 1 and y_idx % 2 == 1):
            #hatching on horizontally
            next_x = seq[x_idx+1]
            hatch_num = 25
            hatch_y_delta = y_delta/hatch_num
            for hatch_idx in range(1, hatch_num):
                plt.plot([x, next_x], [y+hatch_y_delta*hatch_idx, y+hatch_y_delta*hatch_idx], 'r-', linewidth=0.5, alpha=.5)
            
            # #hatching vertically
            # next_y = y+y_delta
            # for cur_x in np.arange(x, next_x, 0.005):
            #     plt.plot([cur_x, cur_x], [y, next_y], 'k-', linewidth=0.5)



plt.gcf().set_size_inches(15,10)
plt.gca().set_aspect('equal')
plt.axis('off')
plt.savefig('riley_squeeze_grid.svg')
plt.show()
```

```python
x_cur, y_cur = 0, 0
x_next, y_next = 100, 100
m_delta = 1

linewidth=.5

for _ in range(50):
    plt.plot([x_cur, x_cur], [y_cur, y_next], 'k-', linewidth=linewidth)
    plt.plot([x_cur, x_next], [y_next, y_next], 'k-', linewidth=linewidth)
    plt.plot([x_next, x_next], [y_next, y_cur], 'k-', linewidth=linewidth)
    plt.plot([x_next, x_cur], [y_cur, y_cur], 'k-', linewidth=linewidth)
    x_cur += m_delta
    y_cur += m_delta

    x_next -= m_delta
    y_next -= m_delta

plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(10, 10)
plt.show()
```
