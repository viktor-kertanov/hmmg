```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import permutations, combinations
import random
```

```python
thetas = np.linspace(0, 2*np.pi, 1000)

# Dealing with circles
x_start = 0
y_start = 0

radius = 1
r_small = .06 #for points
x_gap = radius*2.5
y_gap = radius*2.5

row_num_circles = 5
col_num_circles = 5


# Dealing with larger circles
for c in range(col_num_circles):
    for r in range(row_num_circles):
        for _ in range(np.random.randint(2,4)):
            plt.plot(np.cos(thetas)*radius + x_gap*r, np.sin(thetas)*radius + c*y_gap, 'k-', linewidth=1)

plt.axis('off')
plt.axis([-radius-1, row_num_circles*2*radius+1, -radius-1, col_num_circles*2*radius+2])
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(14, 14)
plt.savefig('circle_graph_circle.svg')
plt.show()

# Dealing with random points 
for c in range(col_num_circles):
    for r in range(row_num_circles):
        num_points = 8
        random_angles = np.random.choice(thetas, num_points, replace=False)
        random_points = [(np.cos(angle) * radius + x_gap * r, np.sin(angle) * radius + c * y_gap) for angle in random_angles]
        rand_combs = combinations(random_points, r=2)

        for point in random_points:
            for _ in range(np.random.randint(1,2)):
                plt.plot(np.cos(thetas)*r_small + point[0], np.sin(thetas)*r_small + point[1], 'r')

        for combination in rand_combs:
            for rpt in range(np.random.randint(1,2)):
                plt.plot([combination[0][0], combination[1][0]], [combination[0][1], combination[1][1]], 'r-', linewidth=.5)
                
                # if rpt % 2 == 0:
                #     plt.plot([ combination[1][0], combination[0][0]], [combination[1][1], combination[0][1]], 'r-', linewidth=.5)

plt.axis('off')
plt.axis([-radius-1, row_num_circles*2*radius+1, -radius-1, col_num_circles*2*radius+2])
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(14, 14)
plt.savefig('circle_graph_graphs.svg')
plt.show()

# Dealing with colors
color_data = {
    'maroon': 1,
    'darkorange': 2,
    'yellow': 3,
    'forestgreen': 4,
    'deepskyblue': 5,
    'blue': 6,
    'darkviolet': 7,
    'cyan': 8,
    'crimson': 9, 
    'black': 10,
    'dodgerblue': 11,
    'deeppink': 12,
    'teal': 13,
    'olive': 14,
    'fuchsia': 15,
    'khaki': 16
    }

num_colors = 7
colors = list(color_data.keys())[:num_colors]
k = 4

color_combos = list(combinations(colors, r=4))
random.shuffle(color_combos)
# Quarters
t_1 = np.linspace(0, np.pi/2, 1000, endpoint=False)
t_2 = np.linspace(np.pi/2, np.pi, 1000, endpoint=False)
t_3 = np.linspace(np.pi, 3*np.pi/2, 1000, endpoint=False)
t_4 = np.linspace(3*np.pi/2, 2*np.pi, 1000, endpoint=False)
color_thetas = [t_1, t_2, t_3, t_4]

for color in color_data:
    for c in range(col_num_circles):
        for r in range(row_num_circles):
            cirlce_idx = c*row_num_circles+r
            color_combo = color_combos[cirlce_idx]
            if color not in color_combo:
                continue
            color_idx = color_combo.index(color)
            plt.plot(np.cos(color_thetas[color_idx])*radius + x_gap*r, np.sin(color_thetas[color_idx])*radius + c*y_gap, color=color, linewidth=2)

    plt.axis('off')
    plt.axis([-radius-1, row_num_circles*2*radius+1, -radius-1, col_num_circles*2*radius+2])
    plt.gca().set_aspect('equal')
    plt.gcf().set_size_inches(14, 14)
    plt.savefig(f'circle_graph_colors_{color}.svg')
    plt.show()
```

```python
thetas = np.linspace(0, 2*np.pi, 1000)

# Dealing with circles
x_start = 0
y_start = 0

radius = 1
r_small = .06 #for points
x_gap = radius*2.5
y_gap = radius*2.5

row_num_circles = 6
col_num_circles = 5

# Dealing with colors
color_data = {
    'maroon': 1,
    'darkorange': 2,
    'yellow': 3,
    'forestgreen': 4,
    'deepskyblue': 5,
    'blue': 6,
    'darkviolet': 7,
    'cyan': 8,
    'crimson': 9, 
    'black': 10,
    'dodgerblue': 11,
    'deeppink': 12,
    'teal': 13,
    'olive': 14,
    'fuchsia': 15,
    'khaki': 16
    }

num_colors = 16
colors = list(color_data.keys())
k = 8 # nCk

color_combos = list(combinations(colors, r=k))
random.shuffle(color_combos)
# Quarters
t_1 = np.linspace(0, np.pi/2, 1000, endpoint=False)
t_2 = np.linspace(np.pi/2, np.pi, 1000, endpoint=False)
t_3 = np.linspace(np.pi, 3*np.pi/2, 1000, endpoint=False)
t_4 = np.linspace(3*np.pi/2, 2*np.pi, 1000, endpoint=False)
color_thetas = [t_1, t_2, t_3, t_4]

radius_delta = .2

for color in color_data:
    for c in range(col_num_circles):
        for r in range(row_num_circles):
            cirlce_idx = c*row_num_circles+r
            color_combo = color_combos[cirlce_idx]
            if color not in color_combo:
                continue
            
            color_idx = color_combo.index(color)
            if color_idx < 4:
                plt.plot(
                    np.cos(color_thetas[color_idx])*radius + x_gap*r,
                    np.sin(color_thetas[color_idx])*radius + c*y_gap,
                    color=color,
                    linewidth=2
                )
            else:
                color_idx = color_idx % 4
                plt.plot(
                    np.cos(color_thetas[color_idx])*radius*(1-radius_delta) + x_gap*r,
                    np.sin(color_thetas[color_idx])*radius*(1-radius_delta) + c*y_gap,
                    color=color,
                    linewidth=2
                )
            
    plt.axis('off')
    plt.axis([-radius-1, row_num_circles*2*radius+3, -radius-1, col_num_circles*2*radius+3])
    plt.gca().set_aspect('equal')
    plt.gcf().set_size_inches(14, 14)
    plt.savefig(f'circle_graph_colors_{color}.svg')
    plt.show()
```
