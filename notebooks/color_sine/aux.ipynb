```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import permutations
```

```python
x = np.linspace(0, 12*np.pi, 10000)


for idx in range(10):
    y = np.sin(x) + x/10

plt.plot(x, y)
plt.axis('on')
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(15, 15)
plt.show()
```
