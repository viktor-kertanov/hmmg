```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import permutations, combinations
import random
```

```python
thetas = np.linspace(0, 2*np.pi, 1000)

# Dealing with circles
x_start = 0
y_start = 0

radius = 1
r_small = .06 #for points
x_gap = radius*2.5
y_gap = radius*2.5

row_num_circles = 6
col_num_circles = 5

# Dealing with colors
color_data = {
    'maroon': 1,
    'darkorange': 2,
    'yellow': 3,
    'forestgreen': 4,
    'deepskyblue': 5,
    'blue': 6,
    'darkviolet': 7,
    'cyan': 8,
    'crimson': 9, 
    'black': 10,
    'dodgerblue': 11,
    'deeppink': 12,
    'teal': 13,
    'olive': 14,
    'fuchsia': 15,
    'khaki': 16
    }

num_colors = 16
colors = list(color_data.keys())[:14]
print(len(colors))
k = 8 # nCk

color_combos = list(combinations(colors, r=k))
random.shuffle(color_combos)
# Quarters
t_1 = np.linspace(0, np.pi/2, 1000, endpoint=False)
t_2 = np.linspace(np.pi/2, np.pi, 1000, endpoint=False)
t_3 = np.linspace(np.pi, 3*np.pi/2, 1000, endpoint=False)
t_4 = np.linspace(3*np.pi/2, 2*np.pi, 1000, endpoint=False)
color_thetas = [t_1, t_2, t_3, t_4]

radius_delta = .2

for color in color_data:
    for c in range(col_num_circles):
        for r in range(row_num_circles):
            cirlce_idx = c*row_num_circles+r
            color_combo = color_combos[cirlce_idx]
            if color not in color_combo:
                continue
            
            color_idx = color_combo.index(color)
            if color_idx < 4:
                plt.plot(
                    np.cos(color_thetas[color_idx])*radius + x_gap*r,
                    np.sin(color_thetas[color_idx])*radius + c*y_gap,
                    color=color,
                    linewidth=2
                )
            else:
                color_idx = color_idx % 4
                plt.plot(
                    np.cos(color_thetas[color_idx])*radius*(1-radius_delta) + x_gap*r,
                    np.sin(color_thetas[color_idx])*radius*(1-radius_delta) + c*y_gap,
                    color=color,
                    linewidth=2
                )
            
plt.axis('off')
plt.axis([-radius-1, row_num_circles*2*radius+3, -radius-1, col_num_circles*2*radius+3])
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(14, 14)
# plt.savefig(f'circle_graph_colors_{color}.svg')
plt.show()
```

```python
color_data = {
    'maroon': 1,
    'darkorange': 2,
    'yellow': 3,
    'forestgreen': 4,
    'deepskyblue': 5,
    'blue': 6,
    'darkviolet': 7,
    'cyan': 8,
    'crimson': 9, 
    'black': 10,
    'dodgerblue': 11,
    'deeppink': 12,
    'teal': 13,
    'olive': 14,
    'fuchsia': 15,
    'khaki': 16
    }

num_colors = 8
colors = list(color_data.keys())[:num_colors]
k = 4 # nCk

color_combos = list(combinations(colors, r=k))
random.shuffle(color_combos)

color_combos = [list(el) for el in color_combos]
for el in color_combos:
    random.shuffle(el)

# Quarters of Circles
split_circle_parts = 4
theta_delta = 2*np.pi / split_circle_parts

color_thetas = []

for t_part in range(split_circle_parts):
    t_angles = np.linspace(t_part*theta_delta, (t_part+1)*theta_delta, 1000, endpoint=False)
    color_thetas.append(t_angles)

init_radius=.1
radius = .1
radius_delta = .1

for color in colors:
    for idx, col_combo in enumerate(color_combos):
        if color not in col_combo:
            radius+=radius_delta
            continue
        color_idx = col_combo.index(color)
        plt.plot(np.cos(color_thetas[color_idx])*radius, np.sin(color_thetas[color_idx])*radius, color=color, linewidth=2)
        radius+=radius_delta
    
    radius = init_radius


plt.axis('off')
plt.axis([-8, 8, -8, 8])
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(14, 14)
# plt.savefig(f'color_target_{color}.svg')
plt.show()

```
