```python
epsilon=.005
thetas_1 = np.linspace(0, np.pi/2, 1000)
thetas_2 = np.linspace(np.pi/2+epsilon, np.pi, 1000)
thetas_3 = np.linspace(np.pi+epsilon, 3*np.pi/2, 1000)
thetas_4 = np.linspace(3*np.pi/2+epsilon, 2*np.pi, 1000)

rs = np.arange(1, 30)

for r in np.arange(1,30):
    x1 = np.cos(thetas_1)*r+1
    y1 = np.sin(thetas_1)*r+5
    x2 = np.cos(thetas_2)*r
    y2 = np.sin(thetas_2)*r
    x3 = np.cos(thetas_3)*r - 5
    y3 = np.sin(thetas_3)*r - 1
    x4 = np.cos(thetas_4)*r
    y4 = np.sin(thetas_4)*r

    plt.plot([x1[-1], x2[0]], [y1[-1], y2[0]])
    plt.plot([x1[0], x4[-1]], [y1[0], y4[-1]])
    plt.plot(x1, y1, 'k-')
    plt.plot(x2, y2, 'k-')
    plt.plot(x3, y3, 'k-')
    plt.plot(x4, y4, 'k-')

plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(10, 10)
plt.show()
```

```python
epsilon=.005
thetas_1 = np.linspace(0, np.pi/2, 1000)
thetas_2 = np.linspace(np.pi/2+epsilon, np.pi, 1000)
thetas_3 = np.linspace(np.pi+epsilon, 3*np.pi/2, 1000)
thetas_4 = np.linspace(3*np.pi/2+epsilon, 2*np.pi, 1000)

num_circles =40

lw = 1

color_1 = ['maroon', 'darkorange', 'yellow', 'forestgreen', 'olive']

color_2 = ['deepskyblue', 'blue', 'darkviolet', 'cyan', 'red']

color_3 = ['crimson','black', 'dodgerblue', 'deeppink', 'green']


for idx in range(1, num_circles):
    for _ in range(np.random.randint(1,3)):
        r = idx + idx*np.abs(idx-num_circles)/(num_circles)
        r = idx
        r_shift = .4
        x1 = np.cos(thetas_1)*r + 4
        y1 = np.sin(thetas_1)*r+r*r_shift - 4
        x2 = np.cos(thetas_2)*r - 4
        y2 = np.sin(thetas_2)*r+r*r_shift + 12
        x3 = np.cos(thetas_3)*r
        y3 = np.sin(thetas_3)*r+r*r_shift
        x4 = np.cos(thetas_4)*r +4
        y4 = np.sin(thetas_4)*r+r*r_shift -4

        plt.plot(x1, y1, 'k-', linewidth=lw)
        plt.plot(x2, y2, 'k-', linewidth=lw)
        plt.plot(x3, y3, 'k-', linewidth=lw)
        plt.plot(x4, y4, 'k-', linewidth=lw)

plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(12, 12)
plt.axis('off')
plt.axis([-45, 45, -30, 70])
plt.savefig('color_broken_circle/1-black-main.svg')
plt.show()

# First connectors
c1 = np.random.choice(color_1, size=num_circles)
for color in color_1:
    for idx in range(1, num_circles):
        if c1[idx-1] != color:
            continue
        r = idx + idx*np.abs(idx-num_circles)/(num_circles)
        r = idx
        r_shift = .4
        x1 = np.cos(thetas_1)*r + 4
        y1 = np.sin(thetas_1)*r+r*r_shift - 4
        x2 = np.cos(thetas_2)*r - 4
        y2 = np.sin(thetas_2)*r+r*r_shift + 12

        plt.plot([x1[-1], x2[0]], [y1[-1], y2[0]], c1[idx-1], linewidth=lw)

    plt.gca().set_aspect('equal')
    plt.gcf().set_size_inches(12, 12)
    plt.axis('off')
    plt.axis([-45, 45, -30, 70])
    plt.savefig(f'color_broken_circle/1-connectors-{color}.svg')
    plt.show()

# Second connectors
c2 = np.random.choice(color_2, size=num_circles)
for color in color_2:
    for idx in range(1, num_circles):
        if c2[idx-1] != color:
            continue
        r = idx + idx*np.abs(idx-num_circles)/(num_circles)
        r = idx
        r_shift = .4
        x2 = np.cos(thetas_2)*r - 4
        y2 = np.sin(thetas_2)*r+r*r_shift + 12
        x3 = np.cos(thetas_3)*r
        y3 = np.sin(thetas_3)*r+r*r_shift

        plt.plot([x2[-1], x3[0]], [y2[-1], y3[0]], c2[idx-1], linewidth=lw)

    plt.gca().set_aspect('equal')
    plt.gcf().set_size_inches(12, 12)
    plt.axis('off')
    plt.axis([-45, 45, -30, 70])
    plt.savefig(f'color_broken_circle/2-connectors-{color}.svg')
    plt.show()

# Third connectors
c3 = np.random.choice(color_3, size=num_circles)
for color in color_3:
    for idx in range(1, num_circles):
        if c3[idx-1] != color:
            continue
        r = idx + idx*np.abs(idx-num_circles)/(num_circles)
        r = idx
        r_shift = .4
        x3 = np.cos(thetas_3)*r
        y3 = np.sin(thetas_3)*r+r*r_shift
        x4 = np.cos(thetas_4)*r +4
        y4 = np.sin(thetas_4)*r+r*r_shift -4

        plt.plot([x3[-1], x4[0]], [y3[-1], y4[0]], c3[idx-1], linewidth=lw)

    plt.gca().set_aspect('equal')
    plt.gcf().set_size_inches(12, 12)
    plt.axis('off')
    plt.axis([-45, 45, -30, 70])
    plt.savefig(f'color_broken_circle/3-connectors-{color}.svg')
    plt.show()
```
