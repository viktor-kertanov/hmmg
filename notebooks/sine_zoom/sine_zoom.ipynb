```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
space = np.linspace(3*np.pi/2, 2*np.pi, 1000)
num_lines = 300
for l in range(num_lines):
    if l % 5 == 0:
        plt.plot((10*np.sin(space)+l/25)*100, 'k', linewidth=.5)

    plt.axis('square')
    plt.gcf().set_size_inches(15, 15)
    plt.axis('off')
plt.savefig('thin.svg')
plt.show()

for l in range(num_lines):
    if l % 10 == 0:
        plt.plot((10*np.sin(space)+0.5*l/25)*100, 'r', linewidth=1)

    plt.axis('square')
    plt.gcf().set_size_inches(15, 15)
    plt.axis('off') 
plt.savefig('thick.svg')
plt.show()
```

```python
r = 1
x = np.linspace(-r, r, 10000)
y_upper = np.sqrt(r - x**2)
y_lower = -np.sqrt(r-x**2)
max_i = 100

for i in range(1, max_i):
    for _ in range(np.random.randint(1,4)):
        plt.plot(x+2*r, y_upper+i/50, 'k-', linewidth=1)
        plt.plot(x, y_lower+i/50, 'k-', linewidth=1)

plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(14, 14)
plt.axis('off')
plt.savefig('1_gray.svg')
plt.show()

for _ in range(2):
    plt.plot(x+2*r, y_upper + max_i/50, 'r-', linewidth=1)
    plt.plot(x, y_lower + max_i/50, 'r-', linewidth=1)

    plt.plot(x+2*r, y_upper + 1/50, 'r-', linewidth=1)
    plt.plot(x, y_lower + 1/50, 'r-', linewidth=1)

plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(14, 14)
plt.axis('off')
plt.savefig('2_red.svg')
plt.show()
```
