```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y = np.linspace(-7, 7, 10000)

for shift in range(100):
    for minishift in range(10):
        x = 0.02*y**3+0.45*shift+0.01*minishift
        coords = np.column_stack((x, y))
        x_mask = (coords[:, 0] >= 7) & (coords[:, 0] <= 20)
        y_mask = (coords[:, 1] >= -7) & (coords[:, 1] <= 7)
        filtered = coords[x_mask & y_mask]
        plt.plot(filtered[:, 0], filtered[:, 1], 'k-', linewidth=1)



# plt.axis([7,20,-7,7])
plt.axis('off')
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(14,14)
plt.savefig('x_cube.svg')
plt.show()
```

```python
y = np.linspace(-7, 7, 10000)

color_data = {
    'maroon': 1,
    'darkorange': 2,
    'yellow': 3,
    'forestgreen': 4,
    'deepskyblue': 5,
    'blue': 6,
    'darkviolet': 7,
    'cyan': 8,
    'crimson': 9, 
    'black': 10,
    'dodgerblue': 11,
    'deeppink': 12,
    'teal': 13,
    'olive': 14,
    'fuchsia': 15,
    'khaki': 16
    }

colors = list(color_data.keys())[:7]
print(f'Colors: {colors}')

for shift in range(1, 101):
    for minishift in range(10):
        x = 0.03*y**3+0.3*shift+0.01*minishift
        coords = np.column_stack((x, y))
        x_mask = (coords[:, 0] >= 11) & (coords[:, 0] <= 23)
        y_mask = (coords[:, 1] >= -6) & (coords[:, 1] <= 6)
        filtered = coords[x_mask & y_mask]
        plt.plot(filtered[:, 0], filtered[:, 1], 'k-', linewidth=1)
    
    if shift % 25 == 0:
        plt.axis('off')
        plt.axis([11, 23, -6, 6])
        plt.gca().set_aspect('equal')
        plt.gcf().set_size_inches(15,15)
        plt.savefig(f'x_cube/x_cube_{shift}.svg')
        # plt.show()

for color_idx, color in enumerate(colors):
    for shift in range(1, 101):
        if shift % len(colors) != color_idx:
            # print(shift % len(colors))
            continue
        x = 0.03 * y**3 + 0.3*shift + .49
        coords = np.column_stack((x, y))
        x_mask = (coords[:, 0] >= 11) & (coords[:, 0] <= 23)
        y_mask = (coords[:, 1] >= -6) & (coords[:, 1] <= 6)
        filtered = coords[x_mask & y_mask]
        plt.plot(filtered[:, 0], filtered[:, 1], color=color, linewidth=1)

    plt.axis('off')
    plt.axis([11, 23, -6, 6])
    plt.gca().set_aspect('equal')
    plt.gcf().set_size_inches(15,15)
    plt.savefig(f'x_cube/x_cube_color_{color}.svg')
plt.show()
```
