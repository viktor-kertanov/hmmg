```python
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
```

```python
thetas = np.linspace(0, 2*np.pi, 1000)

x = np.cos(thetas)
y = np.sin(thetas)

for row in range(12):
    for col in range(20):
        plt.plot(x+col, y+row, 'k-', linewidth=1)

plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(13, 13)

plt.show()
```

```python
thetas = np.linspace(0, 2*np.pi-.001, 1000)

x = np.cos(thetas)
y = np.sin(thetas)

num_rows = 14
num_cols = 20

for row in range(num_rows):
    for col in range(num_cols):
        random_factor = np.random.rand()
        if random_factor < 0.4:
            continue
        for _ in range(np.random.randint(2, 6)):
            if row == 0 or col == 0 or row == num_rows - 1 or col == num_cols - 1:
                cur_x = x+col
                cur_y = y+row
                coords = np.column_stack((cur_x, cur_y))

                x_mask = (coords[:, 0] > 0) & (coords[:, 0] < num_cols-1)
                y_mask = (coords[:, 1] > 0) & (coords[:, 1] < num_rows-1)
                filtered = coords[x_mask & y_mask]
                if col == 0:
                    sorted_indices = np.argsort(filtered[:, 1])
                    filtered_sorted = filtered[sorted_indices]
                    plt.plot(filtered_sorted[:, 0], filtered_sorted[:, 1], 'k-', linewidth=1)
                else:
                    plt.plot(filtered[:, 0], filtered[:, 1], 'k-', linewidth=1)
                
            else:
                plt.plot(x+col, y+row, 'k-', linewidth=1)

plt.gca().set_aspect('equal')
# plt.axis([-1, 20, -1, 12])
plt.axis('off')
plt.gcf().set_size_inches(13, 13)

filename = datetime.now().strftime('%y%m%d_%H_%M_%S_circles.svg')
plt.savefig(filename)

plt.show()
```

```python
thetas = np.linspace(0, 2*np.pi-.001, 1000)

x = np.cos(thetas)
y = np.sin(thetas)

num_rows = 14
num_cols = 20

fibonacci = [1, 2, 3, 5, 8, 13, 21, 34]

for row in range(num_rows):
    for col in range(num_cols):
        random_factor = np.random.rand()
        if random_factor < 0.5:
            continue
        for _ in range(np.random.choice(int(fibonacci, 1))):
            if row == 0 or col == 0 or row == num_rows - 1 or col == num_cols - 1:
                cur_x = x+col
                cur_y = y+row
                coords = np.column_stack((cur_x, cur_y))

                x_mask = (coords[:, 0] > 0) & (coords[:, 0] < num_cols-1)
                y_mask = (coords[:, 1] > 0) & (coords[:, 1] < num_rows-1)
                filtered = coords[x_mask & y_mask]
                if col == 0:
                    sorted_indices = np.argsort(filtered[:, 1])
                    filtered_sorted = filtered[sorted_indices]
                    plt.plot(filtered_sorted[:, 0], filtered_sorted[:, 1], 'k-', linewidth=1)
                else:
                    plt.plot(filtered[:, 0], filtered[:, 1], 'k-', linewidth=1)
                
            else:
                plt.plot(x+col, y+row, 'k-', linewidth=1)

plt.gca().set_aspect('equal')
# plt.axis([-1, 20, -1, 12])
plt.axis('off')
plt.gcf().set_size_inches(13, 13)

filename = datetime.now().strftime('%y%m%d_%H_%M_%S_circles.svg')
plt.savefig(filename)

plt.show()
```
