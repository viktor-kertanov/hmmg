```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
v = 20
h = 20

x_start = 0
x_stop = v
y_start = 0
y_stop = h

for vertex in range(v+1):
    plt.plot([vertex, vertex], [y_start, y_stop], 'k-')
    if vertex % 10 == 0:
        plt.plot([vertex, vertex], [y_start, y_stop], 'r-')
for horizont in range(h+1):
    plt.plot([x_start, x_stop], [horizont, horizont], 'k-')
    if horizont % 10 == 0:
        plt.plot([x_start, x_stop], [horizont, horizont], 'r-')

plt.axis('off')
plt.gca().set_aspect('equal')
plt.gcf().set_size_inches(13,13)
plt.savefig('20_20_grid.svg')
plt.show()
```
